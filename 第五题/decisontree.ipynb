```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
from sklearn import tree
```

```python
data = ['cold', 'cold', 'warm', 'cold', 'hot', 'hot', 'warm', 'cold', 'warm', 'hot']
values = np.array(data)
print(values)
# integer encode
integer_encoded = LabelEncoder().fit_transform(values)
print(integer_encoded)
# binary encode
integer_encoded = integer_encoded.reshape(len(integer_encoded), 1)
onehot_encoder = OneHotEncoder(sparse=False)
onehot_encoded = onehot_encoder.fit_transform(integer_encoded)
print(onehot_encoded)
```

```python
def Encode(data):
    values = np.array(data)
    integer_encoded = LabelEncoder().fit_transform(values)
    integer_encoded = integer_encoded.reshape(len(integer_encoded), 1)
    onehot_encoder = OneHotEncoder(sparse=False)
    onehot_encoded = onehot_encoder.fit_transform(integer_encoded)
    return onehot_encoded
data = pd.read_excel('C:/Users/Dell/Desktop/data.xlsx')
Age = data['Age']
Age = Encode(Age)
Age = pd.DataFrame(Age, columns = ['Age_1', 'Age_2', 'Age_3'])
print(Age)
Incoming = data['Incoming']
Incoming = Encode(Incoming)
Incoming = pd.DataFrame(Incoming, columns = ['Incoming_1', 'Incoming_2', 'Incoming_3'])
print(Incoming)
Student = data['Student']
Student = Encode(Student)
Student = pd.DataFrame(Student, columns = ['Student_1', 'Student_2'])
print(Student)
Credit = data['Credit Rating']
Credit = Encode(Credit)
Credit = pd.DataFrame(Credit, columns = ['Credit_1', 'Credit_2'])
print(Credit)
Buying = data['Buying']
Buying = np.array(Buying)
Buying = LabelEncoder().fit_transform(Buying)
Buying = pd.DataFrame(Buying, columns = ['Buying'])
print(Buying)
```

```python
df = pd.concat([data[['User id']], Age, Incoming, Student, Credit, Buying], axis = 1)
df
```

```python
X = df.iloc[:, 1:11]
X = np.array(X)
print(X)
Y = df['Buying']
Y = np.array(Y)
print(Y)
clf = tree.DecisionTreeClassifier()
clf.fit(X, Y)
clf.score(X, Y)
```
